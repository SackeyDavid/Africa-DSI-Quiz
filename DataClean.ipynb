```python
import pandas as pd
```

```python
wdm = pd.read_csv("Worldometers_data.csv")

wdm.head()
```

```python
wdm['Total_Cases'] = wdm.Total_Cases.str.replace(',', '')
```

```python
wdm.head()
```

```python
wdm['New_Cases'] = wdm.New_Cases.str.replace(',', '')
wdm['Total_Deaths'] = wdm.Total_Deaths.str.replace(',', '')
wdm['New_Deaths'] = wdm.New_Deaths.str.replace(',', '')
wdm['Recovered'] = wdm.Recovered.str.replace(',', '')
wdm['Active'] = wdm.Active.str.replace(',', '')
wdm['Serious_Critical'] = wdm.Serious_Critical.str.replace(',', '')

_ = wdm.rename(columns = {'Deaths/1 M pop': 'Deaths_1Mpop', 'Tests/1Mpop': 'Tests_1Mpop', 'Tot Cases/1M pop': 'Tot_Cases_M' },
          inplace=True)
```

```python
wdm.dtypes
```

```python

wdm['Tot_Cases_1Mpop'] = wdm.Tot_Cases_1Mpop.str.replace(',', '')
wdm['Deaths_1Mpop'] = wdm.Deaths_1Mpop.str.replace(',', '')
wdm['Tests_1Mpop'] = wdm.Tests_1Mpop.str.replace(',', '')
wdm['Tests'] = wdm.Tests.str.replace(',', '')
```

```python
wdm.head()
```

```python
wdm['New_Cases'].value_counts(dropna=False)
```

```python
wdm.dtypes
```

```python
_ = wdm['New_Cases'].fillna(0, inplace=True)
_ = wdm['Total_Deaths'].fillna(0, inplace=True)
_ = wdm['New_Deaths'].fillna(0, inplace=True)
_ = wdm['Recovered'].fillna(0, inplace=True)
_ = wdm['Active'].fillna(0, inplace=True)
_ = wdm['Serious_Critical'].fillna(0, inplace=True)
_ = wdm['Tot_Cases_1Mpop'].fillna(0, inplace=True)
_ = wdm['Deaths_1Mpop'].fillna(0, inplace=True)
_ = wdm['Tests'].fillna(0, inplace=True)
_ = wdm['Tests_1Mpop'].fillna(0, inplace=True)
```

```python
wdm['Total_Deaths'] = wdm.Total_Deaths.astype(int)
wdm['New_Deaths'] = wdm.New_Deaths.astype(int)
wdm['Recovered'] = wdm.Recovered.astype(int)
wdm['Active'] = wdm.Active.astype(int)
wdm['Serious_Critical'] = wdm.Serious_Critical.astype(int)
wdm['Tot_Cases_1Mpop'] = wdm.Tot_Cases_1Mpop.astype(float)
wdm['Deaths_1Mpop'] = wdm.Deaths_1Mpop.astype(float)
wdm['Tests'] = wdm.Tests.astype(int)
wdm['Tests_1Mpop'] = wdm.Tests_1Mpop.astype(int)
wdm['Tests'] = wdm.Tests.astype(int)
```
